```python
from pymoo.algorithms.moo.nsga2 import NSGA2
from pymoo.optimize import minimize
from molstf_problems.dtlz2_alpha import DTLZ2Alpha, rDTLZ2Alpha, DTLZ2Alpha_H, rDTLZ2Alpha_H
from molstf_problems.rlsf import RLSF, rRLSF, RLSF_H, rRLSF_H
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# set experiment parameters
seed = 42
pop_size = 50
n_gen = 50
n_var = 10 # change number of variables here
problem_class = RLSF # change problem class here
```

```python
problem = problem_class(n_var=n_var)

algorithm = NSGA2(pop_size=pop_size)
res = minimize(problem, algorithm, ('n_gen', n_gen), seed=seed, verbose=False)
```

```python
# plot results
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
ax.plot(res.F[:,0], res.F[:,1], 'o', label="NSGA-II")
ax.plot(problem.pareto_front()[:,0], problem.pareto_front()[:,1], linestyle='dashed', label="Pareto front")
ax.set_title(f"{problem_class.__name__} - {n_var} variables")
    
plt.tight_layout()
plt.legend()
plt.show()
```
